The dataset [commute time](https://github.com/barcelonagse-datascience/academic_files/raw/master/data/commute_time.csv) contains information about the commute time of Bicing users in Barcelona. The information is obtained from interviews carried out in 2023 and 2024 in different city locations.

Each record contains the data obatained from a short interview to bicing users using the bicing service to go from their homes to their workplaces. The interview is carried out at the bicing station after a user has parked her bike.

The dataset contains the following information:
 - year: year of the interview (2023 or 2024)
 - start: neighborhood where the bicing user started her journey
 - s_lat: latitude of the "start" neighborhood
 - s_long: longitude of the "start" neighborhood
 - finish: neighborhood where the bicing user finished her journey
 - f_lat: latitude of the "finish" neighborhood
 - f_long: longitude of the "finish" neighborhood
 - commute_time: length of the commute from the home of the bicing user to her workplace
 - avail: number of empty bike slots available at the destination station
 - weekday: weekday in which the interview was carried out (Mon, Tue, Wed, Thurs or Fri)

We are interested in using this dataset to crete a predictive model for the average commute time between Barcelona's neighborhoods.
